#Text calssification.

```python
from transformers import pipeline
```

```python
classifier = pipeline("text-classification")
```

```python
import pandas as pd
```

```python
text = "I hate Abhijeet"
```

```python
outputs = classifier(text)
print(outputs)
```

#Named entity recognition.

```python
ner_tagger = pipeline("ner")
```

```python
text = "I like to visit Hyderabad"
```

```python
outputs = ner_tagger(text)
```

```python
print(outputs)
```

#Question_answering.

```python
reader = pipeline("question-answering")
```

```python
question = "What does the customer want?"
```

```python
outputs_1 = reader(question=question, context=text)
```

```python
print(outputs_1)
```

#Summarization

```python
summarizer = pipeline("summarization")
```

```python
text = '''Das Ziel ist häufig, Erfahrung und Wissen gemeinschaftlich zu sammeln (kollektive Intelligenz) und in für die Zielgruppe verständlicher Form zu dokumentieren. Die Autoren erarbeiten hierzu gemeinschaftlich Texte, die ggf. durch Fotos oder andere Medien ergänzt werden (kollaboratives Schreiben, E-Collaboration). Ermöglicht wird dies durch ein vereinfachtes Content-Management-System, die sogenannte Wiki-Software oder Wiki-Engine.
Das bekannteste Wiki ist die Online-Enzyklopädie Wikipedia, welche die Wiki-Software MediaWiki einsetzt. Zudem nutzen auch viele Unternehmen Wikis als Teil des Wissensmanagementsystems in ihrem Intranet (standortübergreifend), siehe Enterprise Wiki. Ein einzelnes Dokument, eine Wiki-Seite, kann mit wenigen Klicks (Button Bearbeiten und Button Speichern oder Veröffentlichen) geändert werden. Zu diesem Zweck ist die Wiki-Seite zumeist in Form von Wikitext, einer leicht erlernbaren Auszeichnungssprache, gespeichert.
'''
```

```python
outputs = summarizer(text,max_length=45, clean_up_tokenization_spaces=True)
```

```python
print(outputs)
```

#Translation.

```python
translator = pipeline("translation_de_to_en", model="Helsinki-NLP/opus-mt-de-en")
```

```python
outputs = translator(text,clean_up_tokenization_spaces=True,min_length=100)
```

```python
print(outputs)
```
